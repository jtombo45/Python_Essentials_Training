Write a function "encodeString" that will encode a string like 'AAAAABBBBAAA' as a list of tuples: [('A',5),('B',4),('A',3)] meaning that the string has 
"5 A's, followed by 4 B's, followed by 3 A's"

Then use that function to compress a string containing "ASCII Art" (https://en.wikipedia.org/wiki/ASCII_art)
Write a corresponding function "decodeString" that will take in a list of tuples and print the original string.

```python

def encodeString(stringVal):
    encodedList = []
    prevChar = stringVal[0]
    count = 0
    for char in stringVal:
        if prevChar != char:
            encodedList.append(( prevChar, count))
            count = 0
        prevChar = char
        count = count + 1

    encodedList.append((prevChar,count))
    return encodedList
```

```python
encodedStr = encodeString('AAAAABBBBAAA')
encodedStr
```

```python
def decodeString(encodedList):
    decodedStr = ''
    for item in encodedList:
        decodedStr = decodedStr + item[0] * item[1]
    return decodedStr
```

```python
decodeString(encodedStr)
```
